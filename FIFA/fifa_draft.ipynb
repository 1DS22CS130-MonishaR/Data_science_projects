```python
boxes = {}
for team in clubs['epl']:
    boxes[team] = fifa20[fifa20.club == team]['overall']
```

```python
# boxes
```

```python
# 'transpose' items to parallel key, value lists

labels, data = list([*zip(*boxes.items())])

plt.figure(figsize=(12,8),dpi=200)

plt.title("Overall FIFA ratings by Club: EPL")

plt.boxplot(data,vert=False, showmeans=True, meanprops={'color':'#071a52'}, meanline=True, showfliers=True,
            medianprops={'color':'red','linewidth':2}, flierprops={'markerfacecolor':'green','marker':'D'})


# plt.xlabel("FIFA Overall Rating")
plt.xticks(fontsize=14)
plt.yticks(range(1, len(labels)+1 ),labels,fontsize=10)
plt.show()
```

```python
clean_pace = fifa20[fifa20['pace'].notna()]
```

```python
pace_box = {}
for team in clubs['epl']:
    pace_box[team] = clean_pace[clean_pace.club == team]['pace']
```

```python
labels, data = list([*zip(*pace_box.items())])

plt.figure(figsize=(12,8),dpi=200)

plt.title("Pace of Players by Club: EPL")

plt.boxplot(data,vert=False, showmeans=True, meanprops={'color':'#071a52'}, meanline=True, showfliers=True,
            medianprops={'color':'red','linewidth':2}, flierprops={'markerfacecolor':'green','marker':'D'})


# plt.xlabel("FIFA Overall Rating")
plt.xticks(fontsize=14)
plt.yticks(range(1, len(labels)+1 ),labels,fontsize=10)
plt.xlabel("Pace",fontsize=15)
plt.show()
```

```python
left = []
right = []
for team in clubs['epl']:
    left.append(fifa20.loc[fifa20.club == team]['preferred_foot'].value_counts()[1])
    right.append(fifa20.loc[fifa20.club == team]['preferred_foot'].value_counts()[0])
```

```python
sum(left)
```

```python
left
```

```python
right
```

```python
plt.figure(figsize=(4,4),dpi=150)

colours = ['#f9ed69','#14ffec']
plt.pie([sum(left),sum(right)], labels=['Left Foot','Right Foot'],autopct='%.2f %%', pctdistance = 0.5,colors=colours)
plt.title("Left foot vs Right foot: EPL")
```
